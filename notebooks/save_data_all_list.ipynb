```python
from dotenv import load_dotenv
import os
load_dotenv()
DB_HOST = os.getenv('DATABASE_CONNECTION_HOST')
DB_USER = os.getenv('DATABASE_CONNECTION_USER')
DB_PASSWORD = os.getenv('DATABASE_CONNECTION_PASSWORD')
```

```python
import mysql.connector

mydb = mysql.connector.connect(
  host= DB_HOST,
  user= DB_USER,
  password= DB_PASSWORD
)

print(mydb)
```

```python
cursor = mydb.cursor()
```

```python
import pandas as pd

df_products = pd.read_csv('/home/gustavo/Documentos/selling-pipeline/data/df_list_products.csv')
df_products.head()
```

```python
df_products.columns
```

```python
cursor.execute("""
    CREATE TABLE IF NOT EXISTS dbprodutos.tb_products(
               id varchar(100) PRIMARY KEY,
               Produto VARCHAR(100),
               Categoria_Produto VARCHAR(100),
               Preco FLOAT(10,2),
               Frete FLOAT(10,2),
               Data_Compra DATE,
               Vendedor VARCHAR(100),
               Local_Compra VARCHAR(100),
               Avaliacao_Compra INT,
               Tipo_Pagamento VARCHAR(100),
               Qntd_Parcelas INT,
               Latitude FLOAT(10,2),
               Longitude FLOAT(10,2)
    );
""")
```

```python
data_list = [tuple(row) for i, row in df_products.iterrows()]
data_list
```

```python
sql = "INSERT INTO dbprodutos.tb_products VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s);"

cursor.executemany(sql, data_list)
mydb.commit()
```
